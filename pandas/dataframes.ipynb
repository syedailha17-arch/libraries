# Pandas DataFrame Notes

## What is a DataFrame?
A DataFrame is a 2-dimensional labeled data structure in Pandas. It is like a table with rows and columns, similar to an Excel sheet or SQL table.

It is made up of multiple Series (columns).

---

## Creating a DataFrame

### 1. From Dictionary
A DataFrame can be created using a dictionary where keys become column names and values become column data.

Each column must have equal length.

---

### 2. From List of Lists
A DataFrame can also be created using a list of lists, where each inner list represents a row.

Column names must be defined separately.

---

## Structure of DataFrame
- Rows → observations / records
- Columns → features / variables
- Index → row labels (default is 0, 1, 2...)

---

## Key Concept
- Each column in a DataFrame is a Pandas Series
- DataFrames are used for structured tabular data handling

---

## Why DataFrames are important
- Easy data manipulation
- Supports filtering, sorting, grouping
- Used heavily in Data Science and Machine Learning

---

## Basic Understanding from the Lecture
- Think of DataFrame as a table
- Rows = entries
- Columns = attributes of data
- Most operations are column-based

```python
import pandas as pd

# ==========================
# CREATING DATAFRAMES
# ==========================

# From dictionary
data = {
    "Name": ["Amit", "Sara", "John"],
    "Age": [21, 22, 23],
    "Marks": [85, 90, 88]
}

df = pd.DataFrame(data)
print(df)

# From list of lists
data = [
    ["Amit", 21, 85],
    ["Sara", 22, 90],
    ["John", 23, 88]
]

df = pd.DataFrame(data, columns=["Name", "Age", "Marks"])
print(df)


#
```
